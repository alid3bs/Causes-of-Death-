### 3-Quiz

```python
# build a caculator function which takes 3 argument => (number1, number2, operator(+,-,*, /)) 
#and returns the value based on the operation

def calculator(number1, number2, operator):
    if operator == '+':
        return number1 + number2
    elif operator == '-':
        return number1 - number2
    elif operator == '*':
        return number1 * number2
    elif operator == '/':
        if number2 != 0:
            return number1 / number2
        else:
            return "Error: Division by zero!"


```

```python
calculator(2,0,'/')
```
